```python
import pandas as pd
import os
```

```python
#reading in the temperature csv
temp_file = ("temperature.csv")

temp_df = pd.read_csv(temp_file)

#sanity check
#temp_df.head()
```

```python
#creating a df to be merged
ny_temp_df = temp_df[['datetime', 'New York']]
```

```python
#reading in the humidity csv
humidity_file = ("humidity.csv")

humidity_df = pd.read_csv(humidity_file)

#sanity check
#humidity_df.head()
```

```python
#creating a df to be merged
ny_humidity_df = humidity_df[['datetime', 'New York']]

#sanity check
#ny_humidity_df
```

```python
#first merge
ny_first_df = ny_temp_df.merge(ny_humidity_df, on = 'datetime')

#sanity check
#ny_first_df.head()
```

```python
#rename the columns
ny_first_df = ny_first_df.rename(columns={'New York_x': 'Temperature',
                               'New York_y': 'Humidity'})
```

```python
#reading in the weather description csv
weather_file = ("weather_description.csv")

weather_df = pd.read_csv(weather_file)

#sanity check
#weather_df.head()
```

```python
#creating a df to be merged
ny_weather_df = weather_df[['datetime', 'New York']]

#sanity check
#ny_weather_df.head()
```

```python
#second merge
ny_second_df = ny_first_df.merge(ny_weather_df, on = 'datetime')

#sanity check
#ny_second_df.head()
```

```python
#reading in the wind speed csv
wind_file = ("wind_speed.csv")

wind_df = pd.read_csv(wind_file)

#sanity check
#wind_df.head()
```

```python
#creating a df to be merged
ny_wind_df = wind_df[['datetime', 'New York']]

#sanity check
#ny_wind_df.head()
```

```python
#third merge
ny_third_df = ny_second_df.merge(ny_wind_df, on = 'datetime')

#sanity check
#ny_third_df.head()
```

```python
#rename the columns
ny_third_df = ny_third_df.rename(columns={'New York_x':'Weather Description',
                            'New York_y':'Wind Speed'})
#sanity check
#ny_third_df
```

```python
#dropping the emptys
ny_third_df = ny_third_df.dropna(how='any')

#sanity check
#ny_third_df
```

```python
#convert the temperature to F
temp_data = ny_third_df['Temperature']

converted_temp = []

for item in temp_data:
    f = (float(item) - 273.15) * 1.8 + 32
    converted_temp.append(f)
```

```python
#add the converted temp to the df
ny_third_df['Temperature'] = converted_temp

#round the entire column
ny_third_df['Temperature'] = ny_third_df['Temperature'].round(decimals=2)

#sanity check
#ny_third_df
```

```python
#split the datetime column
ny_third_df[['Date', 'Time']] = ny_third_df.datetime.str.split(" ", expand=True)

#sanity check
#ny_third_df
```

```python
#sanity check
#ny_third_df
```

```python
ny_third_df[['Year', 'Month', 'Day']] = ny_third_df.Date.str.split("-", expand=True)

#sanity check
ny_third_df
```

```python
#order the columns
ny_third_df = ny_third_df[['Date',
                           'Year',
                           'Month',
                           'Day',
                           'Time',
                           'Temperature',
                           'Humidity',
                           'Wind Speed',
                           'Weather Description']]
```

```python
#remove the date column
ny_third_df = ny_third_df.drop(columns = 'Date')

#sanity check
ny_third_df
```

```python
#group by year and month
by_month_df = ny_third_df.groupby(['Year', 'Month']).mean()

#sanity check
by_month_df
```
