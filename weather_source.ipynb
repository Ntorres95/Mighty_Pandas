```python
#create variables for the api url
country = 'US'
postal_code = '10019'
#define the API pull function
def pull(start,end):
    base_df = pd.DataFrame(columns = column_names)
#set the start and end dates for months in the form of a list
    range_months = pd.date_range(start,end,freq='1M')

    #not needed anymore, incorporated in the loop
    #dummy variables to change format to match api url data
    #dummy_months = (range_months).strftime('%Y-%m-%d')
    #dummy_months_end = (range_months+1*range_months.freq).strftime('%Y-%m-%d')

    #for each day from x/x/20xx-x/x/20xx ...
    for day in range_months:
        day_start = (day).strftime('%Y-%m-%d')
        day_end = (day+1*day.freq).strftime('%Y-%m-%d')
        try:
        #make the api call
            r = requests.get(f'https://api.weathersource.com/v1/{w_api_key}/postal_codes/{postal_code},{country}/history.json?period=day&timestamp_between={day_start}T00:00,{day_end}T00:00')
            read = r.json()
            print(f'timestamp_between={day_start}T00:00,{day_end}T00:00,length is:{len(read)}')
            #create dummy to temporarily store the dataframe
            dummy_df = pd.DataFrame(read,index=np.arange(start=0,stop=len(read)))
            base_df = base_df.append(dummy_df,ignore_index=False)
        except KeyError:
            print('Error, nothing found, moving to next')
    return base_df
```

```python
#combine dataframes
weather_df = pd.DataFrame(columns = column_names)
weather_df = weather_df.append([w_2_20_7_20, w_4_19_1_20,w_6_18_3_19,w_8_17_7_18,w_9_16_6_17,w_11_15_8_16,w_1_15_10_15],ignore_index=True)
weather_df
```

```python
#truncate data and rename labels
weather_truncate = weather_df[['timestamp','tempAvg','relHumAvg','snowDepth','windSpdAvg','precip']]
weather = weather_truncate.rename(columns={"timestamp": "Day",
                            "tempAvg": "Temperature (F)",
                            "relHumAvg": "Humitdity (%)",
                            "snowDepth": "Snowfall Depth (in)",
                            "windSpdAvg": "Wind Speed (MPH)",
                            "precip": "Precipitation (in)"})
weather
```

```python
#change days to month by slicing off datetime stopping on month
weather_month = weather
weather_month['Day'] = weather_month['Day'].str.slice(start=0,stop=7)
weather_month = weather_month.rename(columns={'Day':'Month'})
weather_month
```
